```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelBinarizer
import numpy as np

import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("flare_data_compiled.csv")

target_col = 'num_flares'
unsummed_cols = ['num_c_class', 'num_m_class', 'num_x_class']

df = df.dropna()
df[target_col] = df[unsummed_cols].sum(axis = 1)
df = df.drop(unsummed_cols, axis = 1)

feature_names = [col for col in df.columns if col != target_col]

pro_data = df.dropna()
```

```python
pro_data
```

```python
encoder = OneHotEncoder(sparse = True)
enc_data = encoder.fit_transform(pro_data.drop("num_flares", axis = 1))
enc_df = pd.DataFrame.sparse.from_spmatrix(enc_data, columns = encoder.get_feature_names(feature_names))
enc_df
```

```python
enc_df[target_col] = pro_data[target_col]
```

```python
enc_df
```

```python
enc_df.columns
```

```python
enc_df.to_csv("solar_flare.csv", index = False)
```

```python
num_cols = None
one_hot_cols = list(enc_df.columns.drop(target_col))

col_dict = {"continuous": num_cols, "discrete": one_hot_cols}

import json

form = json.dumps(col_dict)
f = open("col_dict.json", "w")
f.write(form)
f.close()

with open('col_dict.json') as file:
    col_dict = json.load(file)
file.close()
col_dict
```
