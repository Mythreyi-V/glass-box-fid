```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = pd.read_csv("breast_cancer_unpro.csv")#, sep = ";")
pro_data = data.dropna()

pro_data = pro_data.drop("id", axis = 1)
pro_data.shape
```

```python
num_cols = pro_data.columns.drop('diagnosis')
one_hot_cols = None

col_dict = {"continuous": list(num_cols), "discrete": one_hot_cols}
```
