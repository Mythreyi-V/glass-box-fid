```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np

data = pd.read_csv("hour.csv")
pro_data = data.dropna()

pro_data = pro_data.drop(['instant', 'casual', 'registered', 'dteday'], axis = 1)
pro_data = pro_data.rename(columns = {"cnt": "total_rental"})
pro_data.shape

num_cols = ["temp", "atemp", "hum", "windspeed", "total_rental"]
cat_cols = [col for col in pro_data.columns if col not in num_cols]

num_data = pro_data[num_cols]
cat_data = pro_data[cat_cols]
```

```python
num_data
```

```python
encoder = OneHotEncoder(sparse = True)
enc_data = encoder.fit_transform(cat_data)
enc_df = pd.DataFrame.sparse.from_spmatrix(enc_data, columns = encoder.get_feature_names(cat_data.columns))
enc_df
```

```python
fin_data = pd.concat([num_data.reset_index(), enc_df.reset_index()], axis = 1, join = 'inner').drop(['index'], axis = 1)
fin_data.to_csv("bike_sharing.csv", index = False)
fin_data
```

```python
num_cols = list(num_data.columns.drop('total_rental'))
one_hot_cols = list(enc_df.columns)

col_dict = {"continuous": list(num_cols), "discrete": one_hot_cols}
```

```python
pro_data.to_csv("bike_sharing.csv", index = False)

import json

form = json.dumps(col_dict)
f = open("col_dict.json", "w")
f.write(form)
f.close()
```
